# Process with Multithread

```python
import threading
from threading import Thread
import time
import concurrent.futures
```

```python
class ThreadWithReturnValue(Thread):
    def __init__(self, group=None, target=None, name=None,
                 args=(), kwargs={}, Verbose=None):
        Thread.__init__(self, group, target, name, args, kwargs)
        self._return = None
    def run(self):
        print(type(self._target))
        if self._target is not None:
            self._return = self._target(*self._args,
                                                **self._kwargs)
    def join(self, *args):
        Thread.join(self, *args)
        return self._return
```

### Normalization

Romove al dirty charaters from text

```python
import string
string.punctuation
from sparknlp.base import *
from sparknlp.annotator import *
from pyspark.sql.functions import *
```

```python
def normalize_text(df):
  df = df.select('citationStringAnnotated',(lower(regexp_replace('citationStringAnnotated', "http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))", "")).alias('cleantext')))
  df = df.select('citationStringAnnotated',(lower(regexp_replace('cleantext', "<.*?>", "")).alias('cleantext')))
  df = df.select('citationStringAnnotated',(lower(regexp_replace('cleantext', "[^a-zA-Z\\s]", "")).alias('cleantext')))
  return df
```

### Tokenization

=> Break the sentences into words

```python
from pyspark.ml.feature import *
from pyspark.sql.functions import col, udf
from pyspark.sql.types import IntegerType
```

```python
def tokenization(df):
  regexTokenizer = RegexTokenizer(inputCol = 'cleantext', outputCol = 'words', pattern = '\\W')
  outputToken = udf(lambda w: len(w), IntegerType())
  df = regexTokenizer.transform(df)
  return df
```

### Remove stop words

```python
from pyspark.ml.feature import StopWordsRemover
```

```python
def remove_stopwords(df):
  remover = StopWordsRemover(inputCol = 'words', outputCol='clean_words')
  df = remover.transform(df)
  return df
```

### Data Preprocessing Function

```python
def data_preprocessing(df):
  df = normalize_text(df)
  df = tokenization(df)
  df = remove_stopwords(df)
  return df
```

### Data Preprocessing with multithread

It takes 584 ms to clean all the dataset

```python
%%time
start_time = time.time()
thread1 = ThreadWithReturnValue(target = data_preprocessing, args=(df1,))
thread2 = ThreadWithReturnValue(target = data_preprocessing, args=(df2,))
thread3 = ThreadWithReturnValue(target = data_preprocessing, args=(df3,))
thread4 = ThreadWithReturnValue(target = data_preprocessing, args = (test_df,))

thread1.start()
thread2.start()
thread3.start()
thread4.start()

df1 = thread1.join()
df2 = thread2.join()
df3 = thread3.join()
test_df = thread4.join()
```

```python
def count(text):
    unique_words = set(text)
    words_list = list(unique_words)
    res_dict = {}
    for word in unique_words:
        res_dict[word] = text.count(word)
    return res_dict
```





#### Compute DTF (Document-Term Frequency)

```python
from operator import add
from pyspark import SparkConf
from pyspark import SparkContext
sc = SparkContext.getOrCreate(SparkConf().setMaster("local[*]"))
```

```python
#document term freq
doc_term_freq = tf.flatMap(lambda x: x.keys()).map(lambda word: (word, 1)).reduceByKey(add)
doc_term_freq.count()
```

```python
# take top 50000 terms
N = 50000
doc_term_freq = sc.parallelize(doc_term_freq.top(N,key=lambda x: x[1]))
doc_term_freq.cache()
```

#### Compute IDF (Inverse-Document Frequency)

```python
num_docs = 4527148
idfs = doc_term_freq.map(lambda x: (x[0], math.log((num_docs / x[1]), 10)))
list_idfs = idfs.collect()
```

```python
len(list_idfs)
```

```python
list_idfs
```

```python
term_ids = idfs.keys().zipWithIndex().collectAsMap()
inverse_term_ids = idfs.keys().zipWithIndex().map(lambda x : (x[1], x[0])).collectAsMap()

# keep a read-only variable cached on each machine rather than shipping a copy of it with tasks. 
broadcasted_terms_ids = sc.broadcast(term_ids).value
reverse_broadcasted_ter_ids = sc.broadcast(inverse_term_ids).value
```

```python
# Test
print(term_ids.get('overview'))
print(inverse_term_ids.get(4))
```

#### Compute TF-IDF

```python
from pyspark.mllib.linalg import SparseVector
```

```python
def get_tfidf(doc):
    l = []
    #total_num_terms = sum(doc.values())
    total_num_terms = 50000
    for key, value in doc.items():
        if key in broadcasted_terms_ids:
            term_id = broadcasted_terms_ids.get(key)
        else:
             continue
        idf = list_idfs[term_id]
        tf = value
        #term tf-idf
        res = idf[1] * tf / total_num_terms
        l.append((term_id, res))
        
    return SparseVector(len(list_idfs),l)
```

```python
tf_idf = tf.map(get_tfidf)
```

# SVD



In the text mining context, SVD (Singular Value Decomposition) is called LSA (Latent Semantic Analysis).

SVD fomular: 
\begin{equation*}
M=USV^T
\end{equation*}

where:


\begin{equation*} U \end{equation*}


* m x n matrix
* each row represents a document
* each column represent a concept

\begin{equation*} S \end{equation*}


*  n x n diagonal matrix
*  each entry represents the strength of a concept

\begin{equation*} V^T \end{equation*}

* n x k matrix
* each row represents a concept
* each column represents a term (word)


```python
import numpy as np
from numpy import *
from pyspark.mllib.linalg import *
from pyspark.mllib.linalg.distributed import *
from sklearn.decomposition import *
```

```python
NUM_CONCEPTS = 15

mat = RowMatrix(test_df.rdd.map(lambda v: Vectors.dense(v.features.toArray())))

svd = mat.computeSVD(NUM_CONCEPTS, True)
```

```python
U =  svd.U
S =  svd.s
VT = svd.V
```

```python
print("U:",U)
print("\nS:\n",S)
print("\nVT:\n",VT)
```

# Find top terms in Concept

```python
# conver VT to array
array_VT = VT.toArray()
print(array_VT)
```

```python
# local
def top_terms_in_concepts(num_results):
    top_terms = []
    
    for compNum in range(num_results):
        comp = array_VT.T[compNum]
        # Sort the weights in the first component, and get the indeces
        indeces = np.argsort(comp).tolist()
        # Reverse the indeces, so we have the largest weights first.
        indeces.reverse()
        terms = []
        weights = []
        for i in indeces[:num_results]:
            term = reverse_broadcasted_ter_ids.get(i) 
            terms.append(term)
        weights = comp[indeces[:10]]
        res = list(zip(terms, weights))
        top_terms.append(res)
        
    return top_terms
```
